# TabICL for regression

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
from tabicl import TabICLRegressor

from tab_utils import make_regression_data
```

## Generate data

```python
# generate random train data
x_train, y_train = make_regression_data(
    num_samples=1000,
    regular_grid=False,
    random_seed=123,
)

# generate test data on a regular grid
x_test, y_test = make_regression_data(
    num_samples=1001,
    regular_grid=True,
    random_seed=456,
)

print(f"Features shape.: {x_train.shape}")
print(f"Targets shape: {y_train.shape}")
```

```python
# plot data
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(
    x_train, y_train,
    color=plt.cm.Dark2(7), alpha=0.5,
    edgecolors="none", label="train data"
)
ax.set(xlabel="x", ylabel="y")
ax.set_xlim((0, 10))
ax.legend(loc="upper left")
ax.grid(color="lightgray", linestyle='-')
ax.set_axisbelow(True)
fig.tight_layout()
```

## Test TabICL model

```python
# download checkpoint and initialize model
tabicl = TabICLRegressor()
tabicl.fit(x_train, y_train)

# print(tabicl.model_)
```

```python
# predict via in-context learning
y_pred_mean = tabicl.predict(x_test)

y_pred_quantiles = tabicl.predict(
    x_test,
    output_type="quantiles",
    alphas=[0.1, 0.5, 0.9],  # request median and 10%-90% interval
)

print(f"Mean shape: {y_pred_mean.shape}")
print(f"Quantiles shape: {y_pred_quantiles.shape}")
```

```python
# plot train data and model predictions
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(
    x_train, y_train,
    color=plt.cm.Dark2(7), alpha=0.5,
    edgecolors="none", label="train data",
)
ax.plot(
    x_test, y_pred_mean,
    color=plt.cm.Dark2(2), lw=2.5,
    label="prediction mean",
)
ax.fill_between(
    x_test.flatten(),
    y_pred_quantiles[:, 0],
    y_pred_quantiles[:, 2],
    color=plt.cm.Dark2(2), alpha=0.5,
    label="prediction interval (10%-90%)",
)
ax.set(xlabel="x", ylabel="y")
ax.set_xlim((x_test.min(), x_test.max()))
ax.legend(loc="upper left")
ax.grid(color="gray", alpha=0.2, linestyle="-")
ax.set_axisbelow(True)
fig.tight_layout()
```
